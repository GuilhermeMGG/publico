```python
import matplotlib.pyplot as plt
import random
import pandas as pd
import seaborn as sns
```

```python
dados1 = random.sample(range(100), k=20)
dados2 = random.sample(range(100), k=20)

plt.plot(dados1, dados2) # pyplot gerencia a figura e o eixo
```

```python
dados1 = {

    'Produto':['A', 'B', 'C'],

    'qtde_vendida':[33, 50, 45]

    }

df = pd.DataFrame(dados1)

df.plot(x='Produto', y='qtde_vendida', kind='bar', figsize=[3,3], fontsize=7)

df.plot(x='Produto', y='qtde_vendida', kind='pie', figsize=[3,3])

df.plot(x='Produto', y='qtde_vendida', kind='line', figsize=[3,3]) 
```

```python
sns.set(style='whitegrid') # opes: darkgrid, whitegrid, dark, white, ticks

df_tips = sns.load_dataset('tips')

fig, ax = plt.subplots(1, 3, figsize=(15, 5))

sns.barplot(data=df_tips, x='sex', y='total_bill', ax=ax[0])
sns.barplot(data=df_tips, x='sex', y='total_bill', ax=ax[1], estimator=sum)
sns.barplot(data=df_tips, x='sex', y='total_bill', ax=ax[2], estimator=len)
```

```python
sns.set(style='whitegrid')

df = sns.load_dataset('tips') 

plt.figure(figsize=(4, 2.5))
sns.barplot(x='time', y='total_bill', data=df, estimator=sum, errorbar=None, palette='Set2', legend=False)
plt.xlabel('Perodo (Time)')
plt.ylabel('Total de Gastos')
plt.title('Total de Gastos por Perodo (Almoo ou Jantar)')
plt.show()
```

```python
plt.figure(figsize=(4, 2.5))
sns.barplot(x='time', y='total_bill', data=df)#, #estimator=sum, ci=None, palette=Set2)
plt.xlabel('Perodo (Time)')
plt.ylabel('Mdia de Gastos')
plt.title('Mdia de Gastos por Perodo (Almoo ou Jantar)')
plt.show()
```

```python
# Crie um grfico de barras com o Seaborn para mostrar a mdia de gorjetas por perodo
plt.figure(figsize=(4, 2.5))
sns.barplot(x='time', y='total_bill', data=df, palette='Set3')

plt.xlabel('Perodo (Time)')
plt.ylabel('Mdia da Gorjeta')
plt.title('Mdia da Gorjeta por Perodo (Almoo ou Jantar)')
plt.show()
```
